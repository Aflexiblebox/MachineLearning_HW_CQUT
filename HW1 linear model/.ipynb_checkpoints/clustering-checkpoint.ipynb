```python
import pandas as pd
from sklearn.mixture import BayesianGaussianMixture
from sklearn.preprocessing import OneHotEncoder
import numpy as np
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA

csv_file = "Wholesale customers data.csv"
csv_data = pd.read_csv(csv_file, low_memory = False)#防止弹出警告
csv_df = pd.DataFrame(csv_data)
```

```python
data_slice = np.array(csv_data.iloc[:,2:7])
channel = np.array(csv_data.iloc[:,0]).reshape(-1,1)
region = np.array(csv_data.iloc[:,1]).reshape(-1,1)

enc_channel = OneHotEncoder()
enc_channel.fit(channel)
channel = enc_channel.transform(channel).toarray()
```

```python
enc_region = OneHotEncoder()
enc_region.fit(channel)
region = enc_region.transform(channel).toarray()

data = np.concatenate((channel,region,data_slice),axis=1)
X_train, X_test = train_test_split(data, test_size=0.33, random_state=42)

pca = PCA(n_components=2)
pca.fit(data)


n_components = 12
mixture_model = BayesianGaussianMixture(
    n_components=n_components,
    random_state=5,  # control the pseudo-random initialization
    weight_concentration_prior_type='dirichlet_distribution',
    weight_concentration_prior=1.0,  # parameter of the Dirichlet component prior
    max_iter=2000,  # choose this to be big in case it takes a long time to fit
)
mixture_model.fit(X_train)
y_train = mixture_model.predict(X_train)
```

```python
score = mixture_model.score(X_train)
y_val = mixture_model.predict(X_test)

X_test_vis = pca.transform(X_test)
print(X_test_vis.shape)
```

```python
plt.scatter(X_test_vis[:,0],X_test_vis[:,1],c=y_val)

print("n_component:",n_components)
print("score:",score)
```
